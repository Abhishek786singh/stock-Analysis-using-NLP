```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Desktop/Combined_News_DJIA.csv",encoding="ISO-8859-1")
```

```python
df.head()
```

## dividing the data into train and test

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

## removing spaces

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
## Beacuse their is to much colunms converting it into numerical name
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

## converting the heading into lowercase

```python
# because during analysis if we had same data with starting with 
#lower case and another with upper case causes problems
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

## join data

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

## for analysis 

```python
# for analysis it important that it shuld be in one paragraph
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

##  implement BAG OF WORDS

```python

countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
countvector
```

##implement RandomForest Classifier

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

## predict for the dataset

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

## check accuracy

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
